# Data Visualization with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Data Loading

```python
x = np.linspace(0, 5, 11)
y = x**2
z = np.linspace(-0.75, 1., 100)
```

## Basic Concepts
### Basic Commands

```python
plt.plot(x, y, 'r');
plt.xlabel('X Axis Title');
plt.ylabel('Y Axis Title');
plt.title('Plot Title');
plt.show()
```

### Multiplots on Same Canvas

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--');
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-');
```

### Object Oriented Method

```python
fig = plt.figure()
ax_1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax_2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

ax_1.plot(x, y, 'b');
ax_1.set_xlabel('X_1 Axis Title');
ax_1.set_ylabel('Y_1 Axis Title');
ax_1.set_title('Plot_1 Title');

ax_2.plot(y, x, 'r')
ax_2.set_xlabel('X_2 Axis Title');
ax_2.set_ylabel('Y_2 Axis Title');
ax_2.set_title('Plot_2 Title');
```

### Subplots

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4), dpi=100)

for ax in axes:
    ax.plot(x, y, 'g');
    ax.set_xlabel('X Axis Title');
    ax.set_ylabel('Y Axis Title');
    ax.set_title('Plot Title');

plt.tight_layout()
```

### Legends, Labels and Titles

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label='x**2');
ax.plot(x, x**3, label='x**3');
ax.set_xlabel('X Axis Title');
ax.set_ylabel('Y Axis Title');
ax.set_title('Plot Title');
ax.legend(loc=0);
```

### Colors, Linewidths and Linetypes

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x + 1, color='red', linewidth=0.25);
ax.plot(x, x + 2, color='red', linewidth=0.50);
ax.plot(x, x + 3, color='red', linewidth=1.00);
ax.plot(x, x + 4, color='red', linewidth=2.00);

ax.plot(x, x + 5, color='green', lw=3, linestyle='-');
ax.plot(x, x + 6, color='green', lw=3, ls='-.');
ax.plot(x, x + 7, color='green', lw=3, ls=':');

line, = ax.plot(x, x + 8, color='black', lw=1.50);
line.set_dashes([5, 10, 15, 10])

ax.plot(x, x + 9, color='blue', lw=3, ls='-', marker='+');
ax.plot(x, x + 10, color='blue', lw=3, ls='--', marker='o');
ax.plot(x, x + 11, color='blue', lw=3, ls='-', marker='s');
ax.plot(x, x + 12, color='blue', lw=3, ls='--', marker='1');

ax.plot(x, x + 13, color='purple', lw=1, ls='-', marker='o', markersize=2);
ax.plot(x, x + 14, color='purple', lw=1, ls='-', marker='o', markersize=4);
ax.plot(x, x + 15, color='purple', lw=1, ls='-', marker='o', markersize=8, markerfacecolor='red');
ax.plot(x, x + 16, color='purple', lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green');
```

### Plot Range

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(14, 4))

axes[0].plot(x, x**2, x, x**3);
axes[0].set_title('Default Axes Range');

axes[1].plot(x, x**2, x, x**3);
axes[1].axis('tight');
axes[1].set_title('Tight Axes');

axes[2].plot(x, x**2, x, x**3);
axes[2].set_ylim([0, 60]);
axes[2].set_xlim([2, 5]);
axes[2].set_title('Custom Axes Range');
```

## Advanced Concepts
### Logarithmic Scale

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))
      
axes[0].plot(x, x**2, x, np.exp(x));
axes[0].set_title('Normal Scale');

axes[1].plot(x, x**2, x, np.exp(x));
axes[1].set_yscale('log');
axes[1].set_title('Logarithmic Scale');
```

### Ticks

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(x, x**2, x, x**3, lw=2);

xticks = [1, 2, 3, 4, 5]
ax.set_xticks(xticks);
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18);

yticks = [0, 50, 100, 150]
ax.set_yticks(yticks);
ax.set_yticklabels(['$%.1f$' % y for y in yticks], fontsize=18);
```

### Scientific Notation

```python
from matplotlib import ticker

fig, ax = plt.subplots(figsize=(10, 4))
      
ax.plot(x, x**2, x, np.exp(x));
ax.set_xlabel('X Axis Title');
ax.set_ylabel('Y Axis Title');
ax.set_title('Scientific Notation');
ax.set_yticks([0, 50, 100, 150]);

formatter = ticker.ScalarFormatter(useMathText=True);
formatter.set_scientific(True);
formatter.set_powerlimits((-1, 1));
ax.yaxis.set_major_formatter(formatter);
```

### Axis Number and Axis Label Spacing

```python
plt.rcParams['xtick.major.pad'] = 5
plt.rcParams['ytick.major.pad'] = 5

fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(x, x**2, x, np.exp(x));
ax.set_xlabel('X Axis Title');
ax.set_ylabel('Y Axis Title');
ax.set_title('Label and Axis Spacing');
ax.set_yticks([0, 50, 100, 150]);

ax.xaxis.labelpad = 5
ax.yaxis.labelpad = 5
```

### Axis Position

```python
fig, ax = plt.subplots(figsize=(10, 4))
      
ax.plot(x, x**2, x, np.exp(x));
ax.set_xlabel('X Axis Title');
ax.set_ylabel('Y Axis Title');
ax.set_title('Plot Title');
ax.set_yticks([0, 50, 100, 150]);

fig.subplots_adjust(left=0.15, right=.9, bottom=0.1, top=0.9);
```

### Axis Grid

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

axes[0].plot(x, x**2, x, x**3, lw=2);
axes[0].grid(True);
axes[0].set_title('Default Grid');

axes[1].plot(x, x**2, x, x**3, lw=2);
axes[1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5);
axes[1].set_title('Custom Grid');
```

### Axis Spines

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.spines['bottom'].set_color('blue');
ax.spines['top'].set_color('blue');
ax.spines['left'].set_color('red');
ax.spines['left'].set_linewidth(2);
ax.spines['right'].set_color('none');
ax.yaxis.tick_left();
```

### Twin Axes

```python
fig, ax_1 = plt.subplots(figsize=(10,4))

ax_1.plot(x, x**2, lw=2, color='blue');
ax_1.set_ylabel(r'Area $(m^2)$', fontsize=18, color='blue');
for label in ax_1.get_yticklabels():
    label.set_color('blue');
    
ax_2 = ax_1.twinx();
ax_2.plot(x, x**3, lw=2, color='red');
ax_2.set_ylabel(r'Volume $(m^3)$', fontsize=18, color='red');
for label in ax_2.get_yticklabels():
    label.set_color('red');
```

### Text Annotation

```python
fig, ax = plt.subplots(figsize=(10,4))

ax.plot(z, z**2, z, z**3);
ax.text(0.15, 0.2, r'$y=z^2$', fontsize=20, color='blue');
ax.text(0.65, 0.1, r'$y=z^3$', fontsize=20, color='green');
```

### Advanced Subplots
#### subplot2grid

```python
fig = plt.figure()
ax_1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax_2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax_3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax_4 = plt.subplot2grid((3,3), (2,0))
ax_5 = plt.subplot2grid((3,3), (2,1))
fig.tight_layout()
```

#### gridspec

```python
import matplotlib.gridspec as gridspec

fig = plt.figure()

gs = gridspec.GridSpec(2, 3, height_ratios=[2,1], width_ratios=[1,2,1])
for g in gs:
    ax = fig.add_subplot(g)
    
fig.tight_layout()
```

### Colormap and Contour Figures

```python
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2 * phi_p)
```

```python
phi_m = np.linspace(0, 2 * np.pi, 100)
phi_p = np.linspace(0, 2 * np.pi, 100)
X, Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T
```

#### pcolor

```python
fig, ax = plt.subplots()

p = ax.pcolor(X / (2 * np.pi), Y / (2 * np.pi), Z, cmap='RdBu', vmin=abs(Z).min(), vmax=abs(Z).max());
cb = fig.colorbar(p, ax=ax);
```

#### imshow

```python
fig, ax = plt.subplots()

im = ax.imshow(Z, cmap='RdBu', vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1]);
im.set_interpolation('bilinear');
cb = fig.colorbar(im, ax=ax);
```

#### contour

```python
fig, ax = plt.subplots()

cnt = ax.contour(Z, cmap='RdBu', vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1]);
```

### 3D figures

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

#### plot_surface

```python
fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot(1, 2, 1, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0);

ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='coolwarm', linewidth=0, antialiased=False);

cb = fig.colorbar(p, shrink=0.5);
```

#### plot_wireframe

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1, 1, 1, projection='3d')

p = ax.plot_wireframe(X, Y, Z, rstride=4, cstride=4);
```

#### contour

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1, 1, 1, projection='3d')

ax.plot_surface(X, Y, Z, rstride=4, cstride=4, alpha=0.25);
cset = ax.contour(X, Y, Z, zdir='z', offset=-np.pi, cmap='coolwarm');
cset = ax.contour(X, Y, Z, zdir='x', offset=-np.pi, cmap='coolwarm');
cset = ax.contour(X, Y, Z, zdir='y', offset=3 * np.pi, cmap='coolwarm');

ax.set_xlim3d(-np.pi, 2 * np.pi);
ax.set_ylim3d(0, 3 * np.pi);
ax.set_zlim3d(-np.pi, 2 * np.pi);
```
